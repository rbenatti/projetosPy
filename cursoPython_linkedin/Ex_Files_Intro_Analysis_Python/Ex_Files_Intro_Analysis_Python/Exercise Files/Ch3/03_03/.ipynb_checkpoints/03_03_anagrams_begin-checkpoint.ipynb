```python
wordclean = [word.strip().lower() for word in open('words','r')]
```

```python
wordclean[:10]
```

```python
import collections
```

```python
wordclean = sorted(list(set([word.strip().lower() for word in open('words','r')])))
```

```python
def signature(word):
    return ''.join(sorted(word))
```

```python
words_bysig = collections.defaultdict(list)

for word in wordclean:
    words_bysig[signature(word)].append(word)
```

```python
def anagram_fast(myword):
    return words_bysig[signature(myword)]
```

```python
anagrams_all = {word: anagram_fast(word) for word in wordclean if len(anagram_fast(word)) > 1}
```

```python
anagrams_all
```

```python
len(anagrams_all)
```

```python
# Criar um dict pelo tamanho das palavras
words_bylength = collections.defaultdict(list)

for word in wordclean:
    words_bylength[len(word)].append(word)
```

```python
words_bylength
```

```python
for chave in words_bylength.keys():
    print("size ",chave)
```

```python
# anagramas por tamanho
anagrams_bylength = {}

for lenght, words in words_bylength.items():
    anagrams_bylength[lenght] = {word: anagram_fast(word) for word in words if len(anagram_fast(word)) > 1}
```

```python
anagrams_bylength
```

```python
for lenght, words in words_bylength.items():
    anagrams_bylength[lenght] = {word: len(anagram_fast(word))-1 for word in words if len(anagram_fast(word)) > 1}
```

```python
anagrams_bylength
```

```python
anagrams_bylength = {}

for lenght, words in words_bylength.items():
    anagrams_bylength[lenght] = sum(len(anagram_fast(word))-1 for word in words if len(anagram_fast(word)) > 1)/2
```
